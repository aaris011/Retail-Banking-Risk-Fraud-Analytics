## Memory Optimization

```python
transaction_memory_before = transaction.memory_usage(deep=True).sum() / 1024**2
identity_memory_before = identity.memory_usage(deep=True).sum() / 1024**2

print(f"Transaction Memory Before: {transaction_memory_before:.2f} MB")
print(f"Identity Memory Before: {identity_memory_before:.2f} MB")
```

```python
def reduce_memory_usage(df):

    for col in df.columns:

        col_type = df[col].dtype

        if col_type != object:

            c_min = df[col].min()
            c_max = df[col].max()

            if str(col_type)[:3] == "int":

                if c_min >= np.iinfo(np.int8).min and c_max <= np.iinfo(np.int8).max:
                    df[col] = df[col].astype(np.int8)

                elif c_min >= np.iinfo(np.int16).min and c_max <= np.iinfo(np.int16).max:
                    df[col] = df[col].astype(np.int16)

                elif c_min >= np.iinfo(np.int32).min and c_max <= np.iinfo(np.int32).max:
                    df[col] = df[col].astype(np.int32)

            else:

                df[col] = df[col].astype(np.float32)

    return df
```

```python
transaction = reduce_memory_usage(transaction)
identity = reduce_memory_usage(identity)
```

```python
transaction_memory_after = transaction.memory_usage(deep=True).sum() / 1024**2
identity_memory_after = identity.memory_usage(deep=True).sum() / 1024**2

print(f"Transaction Memory After: {transaction_memory_after:.2f} MB")
print(f"Identity Memory After: {identity_memory_after:.2f} MB")
```

```python
transaction_saved = (
    (transaction_memory_before - transaction_memory_after)
    / transaction_memory_before
) * 100

identity_saved = (
    (identity_memory_before - identity_memory_after)
    / identity_memory_before
) * 100

print(f"Transaction Memory Reduced: {transaction_saved:.2f}%")
print(f"Identity Memory Reduced: {identity_saved:.2f}%")
```

### Memory Optimization Summary

- Successfully optimized numerical data types.
- Reduced dataset memory usage significantly.
- Optimized datasets will improve processing speed during analysis and model training.

## Missing Value Handling Strategy

### Transaction Dataset

- Columns with >80% missing values will be considered for removal.
- Columns with 30–80% missing values will be evaluated individually.
- Columns with <30% missing values will be imputed using appropriate techniques.

### Identity Dataset

- Very high missing columns will be dropped if they provide little information.
- Moderate missing columns will be imputed based on data type.
- Low missing columns will be filled using statistical methods or mode.

### Goal

Create a clean dataset while preserving as much useful information as possible for machine learning.

```python
# Drop columns with more than 80% missing values

transaction = transaction.drop(columns=high_missing.index)

identity = identity.drop(columns=high_missing_id.index)

print("Transaction Shape:", transaction.shape)
print("Identity Shape:", identity.shape)
```

```python
print("Transaction Missing Values:")
print(transaction.isnull().sum().sort_values(ascending=False).head(10))

print("\nIdentity Missing Values:")
print(identity.isnull().sum().sort_values(ascending=False).head(10))
```

```python
# Fill numerical missing values with median

num_cols_transaction = transaction.select_dtypes(include=["int64", "float64"]).columns
num_cols_identity = identity.select_dtypes(include=["int64", "float64"]).columns

for col in num_cols_transaction:
    transaction[col] = transaction[col].fillna(transaction[col].median())

for col in num_cols_identity:
    identity[col] = identity[col].fillna(identity[col].median())

print("Numerical missing values filled successfully.")
```

```python
# Fill categorical missing values with mode

cat_cols_transaction = transaction.select_dtypes(include=["object"]).columns
cat_cols_identity = identity.select_dtypes(include=["object"]).columns

for col in cat_cols_transaction:
    transaction[col] = transaction[col].fillna(transaction[col].mode()[0])

for col in cat_cols_identity:
    identity[col] = identity[col].fillna(identity[col].mode()[0])

print("Categorical missing values filled successfully.")
```
